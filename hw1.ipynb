```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

x = np.linspace(-1, 250, 100)
y = np.tan(x)

x1 = np.linspace(1, 250, 100)
y1 = np.cos(x1)

for i in range(1,100):
    x2 = np.linspace(-3, 250, i)
    y2 = np.cos(x2)

for j in range(7,19):
    x3 = np.linspace(-1, 250, j)
    y3 = np.cos(x3)
    
x4 = np.linspace(1, 250, 500)
y4 = np.tan(np.cos(x4))

x5 = np.linspace(0, 250, 1000)
y5 = np.cos(x5)

plt.xkcd()
plt.plot(x,y, "r")
plt.plot(x1, y1, "g")
plt.plot(np.sin(x2), np.cos(y2), "b")
plt.plot(np.cos(x3), np.tan(x3))
plt.plot(x4, y4)
plt.plot(x5, y5)
```
